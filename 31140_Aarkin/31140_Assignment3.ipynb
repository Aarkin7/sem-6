#### checking for missing values

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Salary'].fillna(df['Salary'].mean(),inplace=True)
```

```python
mode = df['College'].mode()[0]
df['College'].fillna(mode,inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Weight'].fillna(df['Weight'].median(),inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

### Statistical Analysis

```python
df['Age'].describe()
```

```python
df['Age'].value_counts()
```

```python
bins= [11,21,31,41]
labels = ['11 to 20','21 to 30','31 to 40']
df['ageGroup'] = pd.cut(df['Age'], bins=bins,labels=labels, right=False)
```

```python
ageGroup_categories=list(df['ageGroup'].value_counts().index)
ageGroup_categories
```

```python
df['ageGroup'].value_counts()
```

```python
df.head(5)
```

```python
df.groupby(df['ageGroup']).get_group('21 to 30')
```

```python
list_of_salaries_by_ageGroup = list(df.groupby('ageGroup')['Salary'])
list_of_salaries_by_ageGroup
```

```python
df.groupby('ageGroup')['Salary'].mean()
```

```python
df.groupby('ageGroup')['Salary'].median()
```

```python
df.groupby('ageGroup')['Salary'].describe()
```

```python
list_of_colleges_by_ageGroup = list(df.groupby('ageGroup')['College'])
list_of_colleges_by_ageGroup
```
